```python
import pandas as pd
import numpy as np
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

## EV sales(Bus,2W,3W,4W by company)

##### EV Bus sales by manufacturer(2022):-

```python
df_bus_sales = pd.read_csv("EV bus sales by company - 2022.csv")
```

```python
df_bus_sales.head(2)
```

```python
df_bus_sales.drop(['S.NO'],axis=1,inplace=True)
df_bus_sales.rename(columns = {'ELECTRIC(BUS)':'Sales'}, inplace = True)
df_bus_sales['Type']='Bus'
```

```python
df_bus_sales.sort_values('Sales',ascending=False)
```

```python
df_bus=df_bus_sales.sort_values('Sales',ascending=False).head(10)
```

```python
plt.figure(figsize=(14,6))
plt.xticks(rotation = 60)
ax=sns.barplot(x=df_bus['Maker'],y=df_bus['Sales'])
for i in ax.containers:
    ax.bar_label(i,)
```

##### EV 2 wheeler sales by manufacturer(2022):

```python
df_2W_sales = pd.read_csv("EV 2-wheeler sales by company - 2022.csv")
```

```python
df_2W_sales.head(2)
```

```python
df_2W_sales.drop(['S.no'],axis=1,inplace=True)
df_2W_sales.rename(columns = {'ELECTRIC(2-Wheeler)':'Sales'}, inplace = True)
df_2W_sales['Type']='2-Wheeler'
```

```python
df_2W_sales.sort_values('Sales',ascending=False).head()
```

Looks like the datatype of some numbers is set as string

```python
def integer(x):
    x=str(x)
    x=x.replace(',','')
    x=int(x)
    return x
    
```

```python
df_2W_sales['Sales']=df_2W_sales['Sales'].apply(lambda x:integer(x))
```

```python
df_2W=df_2W_sales.sort_values('Sales',ascending=False).head(14)
```

```python
plt.figure(figsize=(14,6))
plt.xticks(rotation = 60)
ax=sns.barplot(x=df_2W['Maker'],y=df_2W['Sales'])
for i in ax.containers:
    ax.bar_label(i,)
```

##### EV 3 wheeler sales by manufacturer(2022):

```python
df_3W_sales = pd.read_csv("EV - 3Wheeler sales by company 2022.csv")
df_3W_sales.head(2)
```

```python
df_3W_sales.drop(['S.no'],axis=1,inplace=True)
df_3W_sales.rename(columns = {'ELECTRIC(3-Wheeler)':'Sales'}, inplace = True)
df_3W_sales['Type']='3-Wheeler'
df_3W_sales.head(1)
```

```python
plt.figure(figsize=(14,6))
plt.xticks(rotation = 60)
ax=sns.barplot(x=df_3W['Maker'],y=df_3W['Sales'])
for i in ax.containers:
    ax.bar_label(i,)
```

##### EV 4 Wheeler Sales by manufacturer

```python
df_4W_sales = pd.read_csv("Ev 4 wheeler sales by maker - 2022.csv")
df_4W_sales.head(2)
```

```python
df_4W_sales.columns
```

```python
df_4W_sales.drop(['S.no','Unnamed: 3', 'Unnamed: 4'],axis=1,inplace=True)
df_4W_sales.rename(columns = {'ELECTRIC(BOV)':'Sales'}, inplace = True)
df_4W_sales['Type']='4-Wheeler'
## To make sure we don't encounter the same problem here, making the sales data integer/
df_4W_sales['Sales']=df_4W_sales['Sales'].apply(lambda x:integer(x))
```

```python
plt.figure(figsize=(14,6))
plt.xticks(rotation = 60)
ax=sns.barplot(x=df_4W['Maker'],y=df_4W['Sales'])
for i in ax.containers:
    ax.bar_label(i,)
```

```python
df= pd.concat([df_bus_sales,df_2W_sales,df_3W_sales,df_4W_sales])
```

```python
plt.figure(figsize=(7,7))
plt.pie(df.groupby('Type').sum()['Sales'],autopct='%1.2f%%')
plt.legend(df.groupby('Type').sum().index)
```

## Charing stations avalable(create a double bar wala plot')

```python
df_stations= pd.read_csv('Charging Station Data - Sheet1.csv')
df_stations.head(1)
```

```python
df_stations.columns
```

```python
df_stations.fillna(0,inplace=True)
```

```python
df_station=df_stations.sort_values('Total',ascending=False)
```

```python
plt.figure(figsize=(18,6))
plt.xticks(rotation = 75)
ax=sns.barplot(x=df_station['Name of State/ UT'],y=df_station['Total'])
for i in ax.containers:
    ax.bar_label(i,)
```

```python
plt.figure(figsize=(16,8))
plt.subplot(1,2,1)
sns.barplot(x=df_stations.sort_values('Charging Stations(Installed as on Nov,2021)',ascending=False).head(10)['Name of State/ UT'],y=df_stations.sort_values('Charging Stations(Installed as on Nov,2021)',ascending=False).head(10)['Charging Stations(Installed as on Nov,2021)'])
plt.xticks(rotation = 75)

plt.subplot(1,2,2)
sns.barplot(x=df_stations.sort_values('Sanctioned to States/ UTs ',ascending=False).head(10)['Name of State/ UT'],y=df_stations.sort_values('Sanctioned to States/ UTs ',ascending=False).head(10)['Sanctioned to States/ UTs '])
plt.xticks(rotation = 75)
plt.show()
```
